### Below we perform some basic operation like Fliping, Rotating, Crop, Blur

```python
from scipy import ndimage  # n dimension
import numpy as np
from matplotlib import pyplot as plt
```

```python
#import image
from skimage import io,img_as_ubyte
img = img_as_ubyte(io.imread('Jaguar.jpg'))
plt.imshow(img)
#img (we will get the array of the image)
```

```python
print(img.shape , img.dtype) #since it has three channels 'RGB' so we get the 3 below
```

```python
flippedLR = np.fliplr(img)  #flippedLeft right
flippedUD = np.flipud(img)  #flipped Up Down
plt.figure(figsize=(8,10))
plt.subplot(3,2,1)
plt.imshow(img)
plt.subplot(3,2,2)
plt.imshow(flippedLR)
plt.subplot(3,2,3)
plt.imshow(flippedUD)
```

### Cropping

```python
img1 = img[500:760,851:1320]
plt.imshow(img1)
```

## Rotation

```python
rotated1 = ndimage.rotate(img,45,reshape=True)   
plt.subplot(1,2,1)
plt.imshow(rotated1)
rotated2 = ndimage.rotate(img,45,reshape=False) #suppose we doesnot fit rotated image in given box
plt.subplot(1,2,2)
plt.imshow(rotated2)
```

## Blur 
We can Blur the image by using gaussian_filtere and decide how much we want to blur the image by fixing the sigma value

Blur means taking avarage of surrounding certain datapoints

```python
plt.imshow(img)     #plot or show original image 
plt.show()
# we can manage blur level by changing up sigma values 
gaussian_filtered = ndimage.gaussian_filter(img, sigma=9)
plt.imshow(gaussian_filtered)
```

### Now we move forward to find the derivatives of the image by using sobel filter with the help of sobel filter we get the derivatives of the image. This is used for 'edge' detection.

```python
sobel_filtered = ndimage.sobel(img,axis=0)
plt.imshow(sobel_filtered)
```
